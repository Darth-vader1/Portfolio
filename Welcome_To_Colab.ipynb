```julia
using Printf
using Plots

f(x) = (1 - cos(x)) / x^2
x = range(-5, stop=5, length=100)
y = f.(x)  # Use broadcasting to apply f to each element of x

xi = [1e-5, 1e-6, 1e-7, 1e-8, 1e-9, 1e-10, 1e-11]
for val in xi
    @printf("f(%g) = %.16f\n", val, f(val))
end
```
